```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0, 7)
y = np.array([0, 0, 31000, 35050, 38906, 43166, 47870])
z = np.array([0, 0, -10000, -13000, -27225, -29948, -32942])
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.set_title('Cash flows')
ax.yaxis.set_label_text('Cash flows')
ax.xaxis.set_label_text('years')
plt.show()
```
